```python
# TASK: Create the two initial sets, staff and directors as shown in the first example above.
# Perform the four mathematical set operations shown, but use the equivalent method calls to achieve the same results. 
# For example:

# Initial sets
staff = {"Pete", "Kelly", "Jon", "Paul", "Sally", "Sue"}
directors = {"Kelly", "Rupert", "Cyril", "Jon"}

staff = staff.union(directors)

common_members = staff.intersection(directors)

staff_only = staff.difference(directors)

unique_members = staff.symmetric_difference(directors)

print("Union:", staff)
print("Intersection:", common_members)
print("Difference:", staff_only)
print("Symmetric Difference:", unique_members)
```

```python
# TASK: Write some code that iterates over the contents of the roots dictionary created within an earlier task.
# For each entry, print the message -

# “The square root of <num> is <sqrt>”
# Where <num> shows the number, and <sqrt> shows the square root of that number.
import math
roots = {n: math.sqrt(n) for n in range(1, 9)}
for num, sqrt_value in roots.items():
    print(f"The square root of {num} is {sqrt_value}")
```
